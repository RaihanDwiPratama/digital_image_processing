```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

1. Buat program image blending dengan 2 foto selfie kalian, gunakan rasio 50:50

```python
img1 = cv2.imread('/content/drive/MyDrive/Footage/Copy of lena1.tif')
img2 = cv2.imread('/content/drive/MyDrive/Footage/Copy of goldhill1.tif')

img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

blend = cv2.addWeighted(img1, 0.5, img2, 0.5, 0.0)

cv2_imshow(blend)
```

2. Unduh satu gambar beresolusi rendah (contohnya bisa berupa gambar yang buram atau terlihat kabur) atau buat gambar beresolusi rendah Anda sendiri. Gunakan OpenCV untuk meningkatkan resolusi gambar tersebut. Anda dapat menggunakan teknik upsampling dengan interpolasi, seperti bilinear atau bicubic. Tampilkan gambar tersebut bersebelahan dengan citra asli!

```python
img3 = cv2.imread('/content/drive/MyDrive/Footage/woman.jpeg')

lowres = cv2.resize(img3, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Citra yang Ditingkatkan Resolusinya')
plt.imshow(cv2.cvtColor(lowres, cv2.COLOR_BGR2RGB))

plt.show()
```

3. Buat program untuk mengetahui kecocokan gambar dengan algoritma Euclidean Distance!

```python
img4 = cv2.imread('/content/drive/MyDrive/Footage/Copy of mervin.jpg')
img5 = cv2.imread('/content/drive/MyDrive/Footage/Copy of rani.jpg')

if img4.shape == img5.shape:
    jarak = np.linalg.norm(img4 - img5)

    print("Jarak Euclidean antara dua gambar:", jarak)

    plt.subplot(1, 2, 1)
    plt.title('Gambar Pertama')
    plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 2, 2)
    plt.title('Gambar Kedua')
    plt.imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))

    plt.show()
else:
    print("Kedua gambar memiliki dimensi yang berbeda.")
```
